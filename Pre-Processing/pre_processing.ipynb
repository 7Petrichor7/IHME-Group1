# Population Part

```python
population = pd.concat([pd.read_csv("./population/IHME_GBD_2019_POP_SYA_" + str(year) + "_Y2021M01D28.CSV") for year in range(1950, 2020)], axis=0)
population.year_id.unique()
```

```python
population.head()
```

```python
population = population[population.location_name.isin(countries)]
population.location_name.nunique()
```

```python
population = population[["measure_name", "location_name", "year_id", "sex_name", "age_group_name", "val"]]
population.head()
```

```python
population = population.rename(columns={'measure_name': 'ds_type_name', 'year_id': 'year'})
population.head()
```

```python
population[population.age == '<1 year']['age'] = '0'
population.head()
```

```python
population = population.rename(columns={'age_group_name': 'age'})
population.head()
```

```python
pop_copy = population.copy()
pop_copy["age_group_name"] = ""
# pop_copy[pop_copy.age.isin(["<1 year", "1", "2", "3", "4"])]["age_group_name"] = "0 to 4"
pop_copy.loc[pop_copy[pop_copy.age.isin(["<1 year", "1", "2", "3", "4"])].index, "age_group_name"] = "0 to 4"
for i in range(5, 94, 5):
    pop_copy.loc[pop_copy[pop_copy.age.isin([str(i), str(i+1), str(i+2), str(i+3), str(i+4)])].index, "age_group_name"] = "" + str(i) + " to " + str(i+4) + ""
pop_copy.loc[pop_copy[pop_copy.age == "95 plus"].index, "age_group_name"] = "95 plus"
pop_copy.head(10)
```

```python
pop_copy = pop_copy.drop(columns=['age'])
pop_copy.sample(10)
```

```python
population = pop_copy.copy()
population.sample(10)
```
